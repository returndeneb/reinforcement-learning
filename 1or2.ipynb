```python
import random
import numpy as np

class Game:
    def __init__(self):
        self.goal = 10
        self.q_table = {s: ([None] if s == self.goal - 1 else [None, None]) for s in range(self.goal)}
        self.state = 0

    def legal_actions(self, state):
        return [a for a in [0, 1] if state + a < self.goal]

    def choose_action(self, state):
        legal = self.legal_actions(state)
        q_vals = self.q_table[state]

        unexplored = [i for i, q in enumerate(q_vals) if q is None]
        if unexplored:
            return random.choice(unexplored) 

        best_q = max(q_vals[a] for a in legal)
        best_actions = [a for a in legal if q_vals[a] == best_q]

        return random.choice(best_actions)

    def update_q(self, state, action, next_state):
        if next_state == self.goal:
            self.q_table[state][action] = 1 
        else:
            next_legal = self.legal_actions(next_state)
            valid_legal = [self.q_table[next_state][a - 1] for a in next_legal if self.q_table[next_state][a - 1] is not None]
            if valid_legal:
                if max(valid_legal) == 1:
                    self.q_table[state][action] = -1
                if sum(np.array(valid_legal)==-1) == 2:
                    self.q_table[state][action] = 1
 
    def play_game(self, verbose=True):
        self.state = 0
        current_player = 1
        if not (None in self.q_table[self.state]):
            print("=======Complete Game========================")
            return False

        while True:

            print(f"현재 상태: {self.state}, Player {(-current_player+3)//2} 차례")
            
            action = self.choose_action(self.state) 
            prev_state = self.state
            self.state += action + 1
            
            self.update_q(prev_state, action, self.state)
            

            print(f"선택한 행동: {action + 1}")
            print(f"업데이트 후 Q-table: {self.q_table}\n")
            

            if self.state == self.goal:
                break

            if not (None in self.q_table[self.state]):
                break

            current_player *= -1
        return True


if __name__ == "__main__":
    game = Game()
    count = 1
    while game.play_game():
        count+=1
    print(f"END update in a {count} runs ")
        
```
