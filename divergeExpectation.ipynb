```python
import random as rd
def onecoin():
    a = 1
    while True:
        if rd.random()>0.5:
            a*=2
        else:
            return a
```

```python
import matplotlib.pyplot as plt


# Plotting in real-time

plt.ion()  # Turn on interactive mode

fig, ax = plt.subplots(figsize=(10, 6))


# Initialize variables

num_trials = 0

total_value = 0

expected_values = []

trial_numbers = []


# Real-time plotting

while num_trials < 1000:

    num_trials += 1

    total_value += onecoin()

    expected_value = total_value / num_trials

    expected_values.append(expected_value)

    trial_numbers.append(num_trials)

    

    # Clear and update plot

    ax.clear()

    ax.plot(trial_numbers, expected_values, label="Expected Value", color="blue")

    ax.axhline(y=2, color='red', linestyle='--', label="Theoretical Value (2)")

    ax.set_xlabel("Number of Trials")

    ax.set_ylabel("Expected Value")

    ax.set_title("Real-Time Calculation of Expected Value")

    ax.legend()

    plt.pause(0.01)



plt.ioff()  # Turn off interactive mode

plt.show()
```

```python
count = 0
value = 0 
for i in range(1000000000):
    count += 1
    value += onecoin()
    if i%3000000==0:
        print(value/count)
```
